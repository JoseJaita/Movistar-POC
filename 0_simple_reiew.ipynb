# Carga de incidencias

```python
mscc = pd.read_csv(
    'MCSS - Fechas de indisponibilidad de Servicio.csv'
)
mscc.columns = [slugify(_) for _ in mscc.columns]

mscc.head(25)
```

```python
mscc.shape
```

```python
def get_date_incident(data, col1, col2):
    return  pd.to_datetime(
        data[col1].astype(str) + ' ' + data[col2].astype(str),
        format='%Y%m %H:%M'
    )

mscc['inicio'] = pd.to_datetime(
    mscc['mesordenado'].astype(str) + ' ' + mscc['hora-inicio'].astype(str)
)
mscc['fin'] = pd.to_datetime(
    mscc['mesordenado'].astype(str)  + ' ' + mscc['hora-fin'].astype(str)
)
```
